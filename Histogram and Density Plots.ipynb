# Histogram and Density Plots using Python

1. Histogram and density plots are a good way to analyze continuous variables. Histograms are generated by bining data to count the number of frequencies in the data set. We can therefore say that the appearance of a histogram depends entirely on the choice of the width of the bin. When analyzing the distribution of data, the bin width is usually set to the default values, but there is a greater chance that the default bin width is not always the most appropriate for any type of data.
2. So it is always very important to try different bin widths to check whether the histogram represents the distribution of the data accurately or not. If the bin width is very small, your histogram will appear as a peak, and if the bin width is very large, it may hide smaller features in the dataset.
3. So, when creating a histogram to analyze the distribution of the dataset, it should be explored across multiple bin widths if you are not satisfied with the default parameters.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
```

```python
data
```

```python
data = pd.DataFrame(data, columns=['x', 'y'])
```

```python
data
```

```python
plt.hist(data["x"], alpha=0.5)
# plt.show()
```

```python
plt.hist(data["y"], alpha=0.5)
# plt.show()
```

# Density Plots

1. Just like histograms where the appearance depends on the bin width we choose, the appearance of density plots is dependent on bandwidth. The bandwidth parameter of density plots behaves the same as the bin width of histograms. If the bandwidth is very small, the density plot will look very sharp, and if the bandwidth is very large, the smaller features in the dataset will disappear.
2. Density plots are created in such a way that the area under the curve is always equal to 1.

```python
df = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
```
