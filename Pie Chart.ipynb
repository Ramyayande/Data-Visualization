
# Pie Chart
A pie chart is created by dividing a circle into two or more sections depending on the number of entities we want to visualize. It is used to analyze the proportion of each entity among all the entities.

# Pie Charts using Python
A pie chart is used to analyze the distribution of data and the proportion of each feature in the data set. Now let’s see how to visualize a simple pie chart using Python, before implementing it on a real dataset. The pyplot.pie() function of the matplotlib library can be used to visualize a pie chart:

```python
import matplotlib.pyplot as plt
```

```python
data = [20, 50, 30, 60]
```

```python
plt.pie(data)
plt.show()
```

In the code above, the pyplot.pie() function takes a list of input values and visualizes a simple pie chart. So it was just a simple pie chart, now let’s see how to visualize pie charts using Python using realtime dataset.

```python
import pandas as pd
```

```python
df = pd.read_json("datanew.json")
```

```python
group_size = [sum(df.positive), sum(df.active), sum(df.cured)]
```

```python
group_labels = ["Positive\n"+str(sum(df.positive)), "Active\n"+str(sum(df.active)), "Cured\n"+str(sum(df.cured))]
```

```python
custom_colors = ["skyblue", "yellowgreen", 'tomato']
```

```python
plt.figure(figsize=(5, 5))
```

```python
plt.pie(group_size, labels=group_labels, colors=custom_colors)
plt.rc('font', size=12)
plt.title("Total Positive, Active, and Cured Cases", fontsize=20)
plt.show()
```

### Now let’s visualize the distribution of top 5 cities with active cases of covid-19 in India:

```python
df.drop(df.tail(1).index, inplace = True) 
```

```python
df1 = df.sort_values(by='active', ascending=False)
```

```python
df3 = df1[:5]
```

```python
states = df3.state_name
```

```python
active = df3.active
```

```python
colours = ["skyblue", "blue", "purple", "yellow", "red"]
```

```python
plt.figure(figsize=(7,7))
```

```python
plt.pie(active, labels=states, colors=colours)
plt.rc('font', size=12)
plt.title("Top 5 Active cases", fontsize=20)
plt.show()
```
